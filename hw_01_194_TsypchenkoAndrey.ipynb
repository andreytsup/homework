__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats as st
welfare = np.array([1000])
df = pd.DataFrame()
for i in range(10**4):
    welfare_change = np.random.choice([-1,1], 1000, p=[0.51, 0.49])
    for g in range(1000): welfare = np.append(welfare, welfare[-1]+welfare_change[g])
    df.insert(i, i, welfare)
    welfare = np.array([1000])
df.insert(0, "mean", np.mean(df, axis=1))
plt.figure(figsize=(10,6))
plt.xlabel("Nomber of day")
plt.ylabel("Wealth")
plt.title("Trajectories of wealth change (Athos)")
plt.plot(df.iloc[:1000, 1:11])
plt.plot(df.iloc[:1000, 0:1], label="mean", linewidth=3, color="black")
plt.ylim (df.iloc[:1000, :11].min().min(), df.iloc[:1000, :11].max().max())
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats as st
df = round(pd.DataFrame({"math": np.random.normal(60, 10, 10**4), "rus": np.random.normal(60, 10, 10**4)}))
df.insert(2, "sum", np.sum(df, axis=1))
df_n = pd.DataFrame({"n": np.arange(80,165, 5)})
df_n.insert(1, "student_count", df_n.n/df_n.n)
df.sort_values(by="sum", inplace=True)
df.reset_index(inplace=True)
for i in range(17):
  ray = np.min(np.where(df["sum"] == df_n.iloc[i, 0])[0])
  df_n.iloc[i, 1] = df.iloc[ray:, 1].corr(df.iloc[ray:, 2])
plt.figure(figsize=(10,6))
plt.plot(df_n.n, df_n.student_count)
plt.title("Correlation between the subjects of passing students depending on passing score")
plt.xlabel("Passing score")
plt.ylabel("Correlation between math and Russian language")
```

Можно предположить, что человеки делают акцент на один предмет в силу чего тащят его и сливают второй, отсюда отрицательная корреляция

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats as st
df = round(pd.DataFrame({"math": np.random.normal(60, 10, 10**4), "rus": np.random.normal(60, 10, 10**4)}))
df.insert(2, "sum", np.sum(df, axis=1))
df_n = pd.DataFrame({"n": np.arange(80,165, 5)})
df_n.insert(1, "student_count", df_n.n/df_n.n)
df.sort_values(by="sum", inplace=True)
df.reset_index(inplace=True)
for i in range(17):
  ray = np.min(np.where(df["sum"] == df_n.iloc[i, 0])[0])
  df_n.iloc[i, 1] = df.iloc[:ray, 1].corr(df.iloc[:ray, 2])
plt.figure(figsize=(10,6))
plt.plot(df_n.n, df_n.student_count)
plt.title("Correlation between the subjects of not passing students depending on passing score")
plt.xlabel("Passing score")
plt.ylabel("Correlation between math and Russian language")
```
